```python
import requests
import bs4
```

```python
request1=requests.get("https://buy.realme.com/in/goods/691")
request1
```

```python
request1.content
```

```python
soup=bs4.BeautifulSoup(request1.text)
soup
```

```python
reviews=soup.findAll('div',{'class':'reviews-primary-p'});
for review in reviews:
    print(review.get_text()+"\n\n")
```

```python
rating=soup.find('div',{'class':''}).get_text()
print(rating)
```
