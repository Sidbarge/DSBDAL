```python
df['Ozone']=df['Ozone'].fillna(df['Ozone'].mean())
```

```python
df['Solar.R']=df['Solar.R'].fillna(df['Solar.R'].mean())
```

```python
df['Humidity']=df['Humidity'].fillna(df['Humidity'].mode()[0])
# Since mode() returns a Series, we select the first element ([0]) from the Series,
# which corresponds to the first mode value. This ensures that we get a single value 
# to fill in for missing values.
```

```python
from sklearn.model_selection import train_test_split
x=df[['Ozone']]
y=df[['Temp']]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=e=0.25)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
model=reg.fit(xtrain,ytrain)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
ypredict=model.predict(xtest)
print("Predicted Data:",ypredict)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))
plt.scatter(df['Ozone'],df['Temp'],color='blue')
plt.plot(xtest,ypredict,color='red')
plt.xlabel('Ozone')
plt.ylabel('Temp')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print(f"MAE:{mean_absolute_error(xtest,ypredict)}")
print(f"MSE:{mean_squared_error(xtest,ypredict)}")
```
